```python
#文章分割のためのモジュール開発
import pandas as pd
import numpy as np
```

```python
def article_to_sentence(df_data):
    #かっこの中か外かの判定処理のためのflag
    flag = 0
    sep_a = []
    for body in df_data["body"][0:10]:
        sentence = ""
        for b in body:
            if(b == "「"):
                flag = 1
            elif(b == "」"):
                flag = 0
            
            if(b == "。" and flag == 0):
                sep_a.append(sentence)
                sentence = ""
                continue

            sentence += b
    
    return sep_a
```

```python
df = pd.read_csv("./reuters_article.csv", dtype={'label': 'str'}, encoding="utf-8")
```

```python
sentence = article_to_sentence(df)
```
